```python
import pandas as pd
```

```python
df = pd.read_excel("./data/全球超市订单数据.xlsx")
```

```python
df.head(2)
```

```python
min_value = df["销售额"].min()
```

```python
max_value = df["销售额"].max()
```

```python
def norm(df):
    df["销售额_norm"] = df["销售额"].apply(
        lambda x : (x-min_value)/(max_value-min_value))
    return df
```

```python
norm(df)
```

```python
df.head(2)
```

```python
df.groupby("细分市场").apply(norm)
```

```python
def topN(data, topn):
    return data.sort_values(by="销售额")[["行 ID", "销售额"]][-topn:]
df.groupby("细分市场").apply(topN, topn=5)
```
