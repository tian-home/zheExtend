```python
import pandas as pd
```

# 0.读取excel数据到dataframe

```python
dpath = "./data/全球超市订单数据.xlsx"
df_raw = pd.read_excel(dpath, nrows=1200)
```

```python
df = df_raw.copy()
df.head(2)
```

# 1.直接赋值的方法

```python
# df.loc[:, "装运方式"] = df["装运方式"].str.replace("级", "")
df["装运方式"] = df["装运方式"].str.replace("级", "等级")
```

```python
df.head(2)
```

```python
def func(x):
    if x > 10000:
        return '优秀'
    elif x > 5000:
        return '一般'
    return '须努力'
df["销售评价"] = df["销售额"].apply(func)
df.head()
```

# 3.assign方法(无)

# 4.按条件分组分别赋值

```python
df["利润率评价"] = ''
df.loc[df["利润"]/df["销售额"]>0.15, "利润率评价"] = "高利润"
df.loc[df["利润"]/df["销售额"]<=0.15, "利润率评价"] = "低利润"
```
