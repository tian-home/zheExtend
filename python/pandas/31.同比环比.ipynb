```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel("./data/全球超市订单数据.xlsx", index_col="订购日期", parse_dates=True)
```

```python
df.head()
```

```python
df = df[["销售额"]].resample("M").mean()
```

```python
df.sort_index(ascending=True, inplace=True)
```

```python
df.head()
```

```python
df.index
```

```python
df.plot()
```

```python
df["订购日期_way1_环比"] = df["销售额"].pct_change(periods=1)
df["订购日期_way1_同比"] = df["销售额"].pct_change(periods=12)
```
