```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/ml-25m/ratings.csv")
```

```python
df.head(3)
```

```python
result = df.groupby("movieId")["rating"].mean()
result.head()
```

```python
type(result)
```

```python
result = df.groupby("movieId")["rating"].agg(
    mean="mean", max="max", min=np.min
)
result.head()
```

```python
result = df.groupby("movieId").agg(
    {"rating": ["mean", "max", np.min]}
)
result.head()
```

```python
result.columns = ["agg_mean", "agg_max", "agg_min"]
result.head()
```

```python
result = df.groupby("movieId").agg(
    rating_mean = ("rating", "mean"),
    rating_max = ("rating", "max"),
    rating_min = ("rating", "min"),
    user_count = ("userId", lambda x : x.nunique())
)
result.head()
```

```python
result = df.groupby("movieId").agg(
    {
        "rating": ["mean", "min", "max"],
        "userId": lambda x :x.nunique()
    }
)
result.head()
```

```python
result.columns = ["a", "b", "c", "d"]
result.head()
```

```python
def func(x):
    return pd.Series({
        "rating_mean": x["rating"].mean(),
        "rating_min": x["rating"].min(),
        "rating_max": x["rating"].max(),
        "user_count": x["userId"].nunique()
    })
result = df.groupby("movieId").apply(func)
result.head()
```
