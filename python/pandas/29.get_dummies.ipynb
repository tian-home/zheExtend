```python
import pandas as pd
```

```python
df_train = pd.read_csv("./data/Titanic-Dataset.csv")
df_train.head()
```

```python
df_train.drop(columns=["Name", "Ticket", "Cabin"], inplace=True)
df_train.head()
```

```python
df_train.info()
```

```python
df_train["Age"] = df_train["Age"].fillna(df_train["Age"].mean())
df_train.info()
```

```python
pd.get_dummies(df_train["Sex"]).head()
```

```python
needcode_cat_columns = ["Pclass", "Sex", "SibSp", "Parch", "Embarked"]
df_coded = pd.get_dummies(
    df_train,
    columns=needcode_cat_columns,
    prefix=needcode_cat_columns,
    dummy_na=True,
    drop_first=True
)
```

```python
df_coded.head()
```

```python
Y = df_coded.pop("Survived")
Y.head()
```

```python
X = df_coded
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver="liblinear")
logreg.fit(X,Y)
```

```python
logreg.score(X,Y)
```
