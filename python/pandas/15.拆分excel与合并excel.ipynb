```python
import pandas as pd
```

```python
df_source = pd.read_csv("./data/Titanic-Dataset.csv")
```

```python
df_source.head()
```

```python
df_source.index
```

```python
df_source.shape
```

```python
total_row_count = df_source.shape[0]
```

```python
total_row_count
```

```python
usernames = ["A", "B", "C", "D", "E", "F"]
```

```python
split_size = total_row_count // len(usernames)
if total_row_count % len(usernames) != 0:
    split_size += 1
split_size
```

```python
df_subs = []
```

```python
for idx, user_name in enumerate(usernames):
    begin = idx*split_size
    end = begin+split_size
    df_sub = df_source.iloc[begin:end]
    df_subs.append((user_name, idx, df_sub))
```

```python
df_subs[0][2]
```

```python
for user_name, idx, df_sub in df_subs:
    file_name = "./data/file_split/{}.xlsx".format(user_name)
    print(file_name)
    df_sub.to_excel(file_name, index=False)
```
