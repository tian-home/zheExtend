```python
import pandas as pd
```

```python
df = pd.read_csv("./data/BIDU历史数据.csv")
```

```python
df.head()
```

```python
df.index
```

```python
df.sort_index(inplace=True)
df.head()
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts
```

```python
line = Line()
line.add_xaxis(df.index.to_list())
line.add_yaxis("开盘价", df["开盘"].round(2).to_list())
line.add_yaxis("收盘价", df["收盘"].round(2).to_list())
line.set_global_opts(
    title_opts=opts.TitleOpts(title="BIDU股票图"),
    tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross")
)
```

```python
line.render_notebook()
```
