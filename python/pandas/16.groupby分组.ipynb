```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_excel("./data/全球超市订单数据.xlsx")
df.head(2)
```

```python
df.groupby("子类别").sum()
```

```python
df.groupby(["细分市场", "子类别"], as_index=False).mean()
```

```python
df.groupby("细分市场").agg([np.sum, np.mean, np.std])
```

```python
df.groupby("细分市场")['销售额'].agg([np.sum, np.mean, np.std])
```

```python
df.groupby("细分市场").agg([np.sum, np.mean, np.std])['销售额']
```

```python
df.groupby("细分市场").agg({"销售额":np.sum, "利润":np.mean})
```

```python
df.groupby("子类别")["销售额"].max()
```

```python
df.groupby("year")["销售额"].plot()
```

```python
df.groupby("year").agg({"销售额":np.sum, "利润":np.sum})
```

```python
df.groupby("year").agg({"销售额":np.sum, "利润":np.sum}).plot()
```
