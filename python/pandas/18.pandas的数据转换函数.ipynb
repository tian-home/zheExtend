```python
import pandas as pd
```

```python
df_BABA = pd.read_csv("./data/BABA历史数据.csv")
df_BABA["公司"]="BABA"
df_BIDU = pd.read_csv("./data/BIDU历史数据.csv")
df_BIDU["公司"]="BIDU"
df_JD = pd.read_csv("./data/JD历史数据.csv")
df_JD["公司"]="JD"
df_PDD = pd.read_csv("./data/PDD历史数据.csv")
df_PDD["公司"]="PDD"
```

```python
df = pd.concat([df_BABA, df_BIDU, df_JD, df_PDD])
```

```python
df
```

```python
df["公司"].unique()
```

```python
dict_company_names = {
    "bidu": "百度",
    "baba": "阿里巴巴",
    "jd": "京东",
    "pdd": "拼多多"
}
```

```python
df["公司中文1"] = df["公司"].str.lower().map(dict_company_names)
```

```python
df.head()
```

```python
df["公司中文2"] = df["公司"].map(lambda x : dict_company_names[x.lower()])
```

```python
df.head()
```

```python
df["公司中文3"] = df["公司"].apply(lambda x : dict_company_names[x.lower()])
```

```python
df.head()
```

```python
df["公司中文4"] = df.apply(lambda x : dict_company_names[x["公司"].lower()], axis=1)
```

```python
df.head()
```
