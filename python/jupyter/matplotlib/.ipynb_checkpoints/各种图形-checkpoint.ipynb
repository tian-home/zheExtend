# 折线图

```python
plt.figure(figsize=(5, 3))

x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
y = []
for i in range(7):
    y.append(random.randint(1, 50))
plt.plot(x, y)
plt.xlabel('星期')
plt.ylabel('活跃度')
plt.title('Python语言活跃度')

for a, b in zip(x, y):
    plt.text(a, b, b, )
```

```python
df_01 = df[df["类别"] == "技术"]
df_01_result = df_01.groupby("year").sum()
df_02 = df[df["类别"] == "家具"]
df_02_result = df_02.groupby("year").sum()
df_03 = df[df["类别"] == "办公用品"]
df_03_result = df_03.groupby("year").sum()
```

```python
plt.plot(df_01_result.index, df_01_result["销售额"], marker='o', label='技术')
plt.plot(df_02_result.index, df_02_result["销售额"], marker='o', label='家具')
plt.plot(df_03_result.index, df_03_result["销售额"], marker='o', label='办公用品')
plt.xlabel("年份")
plt.ylabel("销售额")
plt.title("2012-2015分类别销售额")
plt.legend()
```

# 雷达图

```python
plt.figure(figsize=(5, 5))

x = np.linspace(0, 2*np.pi, 6, endpoint=False)
y = [80, 60, 90, 70, 40, 100]

axes = plt.subplot(111, polar=True)
axes.plot(x, y)
```

# 等高线图

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)

plt.contourf(X, Y, Z)
```

# 3D图

```python
fig = plt.figure(figsize=(5, 3))
x = np.linspace(0, 100, 400)
y = np.sin(x)
z = np.cos(x)

axes = Axes3D(fig)
fig.add_axes(axes)
axes.plot(x, y, z)
```
