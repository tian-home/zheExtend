```python
fig = plt.figure(figsize=(6, 4))

x = np.linspace(-np.pi, np.pi, 30)
y = np.sin(x)

ax1 = plt.subplot(221)
ax1.plot(x, y)

ax2 = plt.subplot(222)
ax2.plot(x, np.cos(x))

ax3 = plt.subplot(223)
ax3.plot(x, np.tan(x))

ax4 = plt.subplot(224)
ax4.plot(x, y)

fig.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 5))

x = np.linspace(-np.pi, np.pi, 30)
y = np.sin(x)

ax1 = plt.subplot(2,2,1)
ax1.plot(x, y)

ax2 = plt.subplot(2,2,2)
ax2.plot(x, y)

ax3 = plt.subplot(2, 1, 2)
ax3.plot(x, np.sin(x*x))
```

```python
x = np.linspace(0, 2*np.pi)

fig, ax = plt.subplots(3, 3)
ax1, ax2, ax3 = ax
ax11, ax12, ax13 = ax1
ax21, ax22, ax23 = ax2
ax31, ax32, ax33 = ax3
fig.set_figwidth(8)
fig.set_figheight(5)

ax11.plot(x, np.sin(x))
ax12.plot(x, np.cos(x))
ax13.plot(x, np.tan(x))
ax21.plot(x, np.sinh(x))
ax22.plot(x, np.cosh(x+x))
ax23.plot(x, np.tanh(x))
ax31.plot(x, np.sin(x + np.cos(x)))
ax32.plot(x, np.sin(x*x + np.cos(x*x)))
ax33.plot(x, np.sin(x * np.cos(x)))
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 5))

x = np.linspace(0, 2*np.pi, 30)
y = np.sin(x)

plt.plot(x, y)

axes1 = plt.axes([0.55, 0.55, 0.3, 0.3])
axes1.plot(x, y, color="g")

axes2 = fig.add_axes([0.18, 0.18, 0.3, 0.3])
axes2.plot(x, y, color="r")
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 10, 100)
axes1 = plt.gca()
axes1.plot(x, np.exp(x), color='r')
axes1.set_xlabel('time')
axes1.set_ylabel('exp')
axes1.tick_params(axis='y', labelcolor="red")

axes2 = axes1.twinx()
axes2.plot(x, np.sin(x), color='b')
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
```

```python
plt.figure(figsize=(6, 4))

x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.legend(['Sin', 'Cos'],
           fontsize=18,
           loc='center',
           ncol=2,
           bbox_to_anchor=[0, 1, 1, 0.2]
          )
```

```python
plt.figure(figsize=(8, 5))

x = np.linspace(0, 2*np.pi, 20)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='r', marker='o', ls='-.', lw=1, label='sinx')

plt.plot(x, y2, color='b', marker='*', ls='--', lw=2, label='cosx')

plt.legend()
```

```python
import random
plt.figure(figsize=(5, 3))

x = np.linspace(1, 10, 10)
# y = np.array([for i in range(11)])
l = []
for i in range(10):
    l.append(random.randint(1, 100))
y = np.array(l)
plt.plot(x, y)
for a, b in zip(x, y):
    print(a, b)
    plt.text(
        x=a,
        y=b,
        s=b
    )
plt.show()
```
