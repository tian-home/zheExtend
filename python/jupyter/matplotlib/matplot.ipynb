```python
0
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny, color="red")
plt.plot(x, cosy)
```

```python
plt.plot(x, siny, color="red")
plt.plot(x, cosy)
plt.xlim(-5, 15)
```

```python
plt.plot(x, siny, color="red")
plt.plot(x, cosy)
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny, color="red")
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2])
```

```python
plt.plot(x, siny, color="red", linestyle="--", label="sin(x)")
plt.plot(x, cosy, label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y label")
plt.legend()
```

```python
plt.plot(x, siny, color="red", linestyle="--", label="sin(x)")
plt.plot(x, cosy, label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y label")
plt.legend()
plt.title("welcome to the ML world!")
```
